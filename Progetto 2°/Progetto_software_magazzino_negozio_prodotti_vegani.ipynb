```python
import json

def check_int(string):
    """
    Controlla la conformità della stringa per numeri interi.
    """

    check_list = []
    valid_list = list(map(str, range(0, 10)))
    for i in string:
        if i in valid_list:
            check_list.append(True)
        else:
            check_list.append(False)
    return False in check_list

def check_float(string):
    """
    Controlla la conformità della stringa per numeri decimali.
    """

    check_list = []
    valid_list = list(map(str, range(0, 10))) + [".", ","]
    for i in string:
        if i in valid_list:
            check_list.append(True)
        else:
            check_list.append(False)
    return False in check_list




warehouse_dict = {}
cmd = None
control_list = ["aggiungi", "elenca", "vendita", "profitti", "aiuto", "chiudi"]

with open("warehouse.json", "r") as warehouse_json:
    warehouse_dict = json.load(warehouse_json)

while cmd != "chiudi":
    cmd = input("Inserisci un comando: ")

    if cmd not in control_list:
        print("Comando non valido\nI comandi disponibili sono i seguenti:\naggiungi: aggiungi un prodotto al magazzino\nelenca: elenca i prodotto in magazzino\nvendita: registra una vendita effettuata\nprofitti: mostra i profitti totali\naiuto: mostra i possibili comandi\nchiudi: esci dal programma")

    elif cmd == "aggiungi":
        name_str = input("Nome del prodotto: ")

        if not name_str in warehouse_dict.keys():
            try:
                quantity_str = input("Quantità del prodotto: ")
                assert(check_int(quantity_str) is False), "Il valore dev'essere un numero intero."
                price_buy_str = input("Prezzo di acquisto: €")
                assert(check_float(price_buy_str) is False), "Il valore dev'essere un numero."
                price_sell_str = input("Prezzo di vendita: €")
                assert(check_float(price_sell_str) is False), "Il valore dev'essere un numero."

            except AssertionError as e:
                print(e)

            else:
                warehouse_dict[name_str] = {"quantity": int(quantity_str), "price_buy": float(price_buy_str), "price_sell": float(price_sell_str), "sold": 0}
                print("AGGIUNTO: %d X %s" % (warehouse_dict[name_str]["quantity"], name_str))
        else:
            try:
                quantity_str = input("Quantità del prodotto: ")
                assert(check_int(quantity_str) is False), "Il valore dev'essere un numero intero."

            except AssertionError as e:
                print(e)

            else:
                sum_int = int(quantity_str) + warehouse_dict[name_str]["quantity"]
                warehouse_dict[name_str]["quantity"] = sum_int

                print("AGGIUNTO: %d X di %s" % (int(quantity_str), name_str))

    elif cmd == "vendita":
        more_str = None
        sold_dict = {}

        while more_str != "no":
            name_str = input("Nome del prodotto: ")

            if not name_str in warehouse_dict.keys():
                print(f"{name_str} non è presente in magazzino.")
                sold_dict[name_str] = [False]

            else:
                sold_dict[name_str] = [True]
                try:
                    quantity_str = input("Quantità del prodotto: ")
                    assert(check_int(quantity_str) is False), "Il valore dev'essere un numero intero."

                except AssertionError as e:
                    print(e)

                else:
                    quantity_int = int(quantity_str)
                    if quantity_int == 0:
                        print("La quantità non può essere zero.")
                    elif quantity_int > warehouse_dict[name_str]["quantity"]:
                        print("%s non è disponibile in questa quantità, quantità attuale: %d." % (name_str, warehouse_dict[name_str]["quantity"]))
                        sold_dict[name_str] = [False]
                    else:
                        sold_dict[name_str] = [True]
                        sub_int = warehouse_dict[name_str]["quantity"] - quantity_int
                        sold_int = warehouse_dict[name_str]["sold"] + quantity_int

                        warehouse_dict[name_str]["quantity"] = sub_int
                        warehouse_dict[name_str]["sold"] = sold_int

                        sold_dict[name_str].append(quantity_int)
                        sold_dict[name_str].append(warehouse_dict[name_str]["price_sell"])

                        more_str = (input("Aggiungere un altro prodotto?(si/no) "))

                        if more_str == "no":
                            tot_list = []
                            print("VENDITA REGISTRATA")

                            keys_list = list(sold_dict.keys())
                            for i in keys_list:
                                if sold_dict[i][0] is True:
                                    tot_list.append(sold_dict[i][1]*sold_dict[i][2])
                                    print("%d X di %s: €%.2f" % (sold_dict[i][1], i, sold_dict[i][2]))
                            print("Totale: €%.2f" % (sum(tot_list)))

    elif cmd == "elenca":
        keys_list = list(warehouse_dict.keys())

        print("PRODOTTO QUANTITA' PREZZO")
        for i in keys_list:
            if warehouse_dict[i]["quantity"] != 0:
                print("%s: %d, € %.2f" % (i, warehouse_dict[i]["quantity"], warehouse_dict[i]["price_sell"]))

    elif cmd == "profitti":
        lordo_list = []
        outgoings_list = []
        keys_list = list(warehouse_dict.keys())

        for i in keys_list:
            lordo_float = warehouse_dict[i]["price_sell"]*warehouse_dict[i]["sold"]
            lordo_list.append(lordo_float)

            outgoings_float = (warehouse_dict[i]["quantity"]+warehouse_dict[i]["sold"])*warehouse_dict[i]["price_buy"]
            outgoings_list.append(outgoings_float)

        tot_netto_float = sum(lordo_list) - sum(outgoings_list)
        print("Profitto: lordo = €%.2f netto = € %.2f" % (sum(lordo_list), tot_netto_float))


    elif cmd == "aiuto":
        print("I comandi disponibili sono i seguenti:\naggiungi: aggiungi un prodotto al magazzino\nelenca: elenca i prodotto in magazzino\nvendita: registra una vendita effettuata\nprofitti: mostra i profitti totali\naiuto: mostra i possibili comandi\nchiudi: esci dal programma")


    else:
        print("Bye bye")


with open("warehouse.json", "w") as warehouse_json:
    json.dump(warehouse_dict, warehouse_json)
```
